# Feature Analysis

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# plt.style.use('bmh')
```

```python
# read in features, parse the date column to date object

df = pd.read_csv('datasets/dengue_features_train.csv', parse_dates=[3])

# target

target = pd.read_csv('datasets/dengue_labels_train.csv')
target.head()

# add total cases into df
df['total_cases'] = target['total_cases']
df.head()
```

```python
df.describe().T
```

```python
target.describe().T
```

### Model Splitting and Cleaning

```python
# create new datasets

df_sj = df[df['city'] == 'sj']
df_iq = df[df['city'] == 'iq']

print('San Juan shape: {}'.format(df_sj.shape))
print('Iquitos shape: {}'.format(df_iq.shape))
```

```python
## FrontFill to Impute into Missing values

df_sj = df_sj.fillna(method = 'ffill')
df_iq = df_iq.fillna(method = 'ffill')
```

## Feature Correlation

```python
# San Juan
sns.heatmap(df_sj.corr()**2)
```

```python
# Iquitos
sns.heatmap(df_iq.corr()**2)
```

### Correlation Plot Conclusions

Total Cases does not correlate to any specific feature individually

** NDVI: ** The satellite imagery score of the vegetation growing in the city does not move together especially in San Juan. The North quadrants and the South quadrants have almost no relation to each other. There is a stronger correlation in Iquitos.

** Temperature: ** The temperature variables (`reanalysis_air_temp_k` to `reanalysis_min_air_temp_k`) are strongly correlated together in San Juan, but not in Iquitos.
